```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from sklearn.linear_model import LinearRegression

# Generate random data
np.random.seed(42)
X = np.random.rand(30) * 10
y = 2 * X + 3 + np.random.randn(30) * 2  # y = 2x + 3 + noise

# Initialize figure
fig, ax = plt.subplots()
ax.scatter(X, y, color='blue', label='Data Points')
line, = ax.plot([], [], 'r-', linewidth=2, label='Regression Line')
ax.set_xlim(0, 10)
ax.set_ylim(0, 25)
ax.legend()

# Gradient Descent-like animation
def init():
    line.set_data([], [])
    return line,

def update(frame):
    # Simulate progressive fitting (simplified)
    temp_slope = 0.1 * frame  # Gradually increase slope
    temp_intercept = 0.5 * frame  # Gradually increase intercept
    x_vals = np.array([0, 10])
    y_vals = temp_slope * x_vals + temp_intercept
    line.set_data(x_vals, y_vals)
    return line,

ani = FuncAnimation(fig, update, frames=20, init_func=init, blit=True, interval=200)
ani.save('linear_regression_fit.mp4', writer='ffmpeg')
plt.show()
```
